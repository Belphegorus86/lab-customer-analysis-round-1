```python
import pandas as pd
```

```python
csv1 = pd.read_csv("https://raw.githubusercontent.com/Belphegorus86/lab-customer-analysis-round-1/master/files_for_lab/csv_files/file1.csv")
csv2 = pd.read_csv("https://raw.githubusercontent.com/Belphegorus86/lab-customer-analysis-round-1/master/files_for_lab/csv_files/file2.csv")
csv3 = pd.read_csv("https://raw.githubusercontent.com/Belphegorus86/lab-customer-analysis-round-1/master/files_for_lab/csv_files/file3.csv")
```

```python
csv1
```

```python
csv1.shape, csv2.shape, csv3.shape
```

```python
csv1 = csv1.drop_duplicates()
csv1
```

```python
csv1.columns = csv1.columns.str.lower().str.replace('state', 'st').str.replace('_', ' ')
csv2.columns = csv2.columns.str.lower().str.replace('state', 'st').str.replace('_', ' ')
csv3.columns = csv3.columns.str.lower().str.replace('state', 'st').str.replace('_', ' ')
```

```python
csv1
```

```python
csv1 = csv1[["customer", "st", "gender", "education", "customer lifetime value", "income", "monthly premium auto", "number of open complaints", "policy type", "vehicle class", "total claim amount"]]
csv2 = csv2[["customer", "st", "gender", "education", "customer lifetime value", "income", "monthly premium auto", "number of open complaints", "policy type", "vehicle class", "total claim amount"]]
csv3 = csv3[["customer", "st", "gender", "education", "customer lifetime value", "income", "monthly premium auto", "number of open complaints", "policy type", "vehicle class", "total claim amount"]]
```

```python
csv1
```

```python
csv2
```

```python
csv3
```

```python
concatcsv = pd.concat([csv1, csv2, csv3], axis=0, ignore_index=True)
```

```python
concatcsv
```

```python
num_cols = concatcsv._get_numeric_data()
```

```python
num_cols
#by omission the rest of the column should be categorical
```

```python
concatcsv = concatcsv.drop(['education', 'number of open complaints'], axis=1)
```

```python
concatcsv
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)

def remove_percent(value):
    return value.rstrip('%') if isinstance(value, str) else value

concatcsv["customer lifetime value"] = concatcsv["customer lifetime value"].apply(remove_percent)
concatcsv["customer lifetime value"] = pd.to_numeric(concatcsv["customer lifetime value"], errors='coerce')

# Multiply by 100
concatcsv["customer lifetime value"] *= 100
```
